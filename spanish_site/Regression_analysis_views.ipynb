```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
questions = pd.DataFrame.from_csv('questions.csv', index_col=None)
questions = questions.fillna(value=0)
questions.head()
```

```python
questions['ViewCount'] = questions['ViewCount'] - questions['ViewCount'].median()
questions['ViewCount'].max()

questions['age'] = (pd.Timestamp('20160314')-pd.to_datetime(questions.Creation,format='%Y-%m-%d %H:%M:%S.%f')).apply(lambda x: x.astype('timedelta64[D]').item().days)
questions['ViewAge'] = questions.ViewCount/questions.age
```

## We take page views (ViewCount) as a reference for profit on StackExchange Network

```python
questions.sort_values(by='Score').groupby('Score').median()[['ViewCount']].plot()
questions.sort_values(by='AnsCount').groupby('AnsCount').median()[['ViewCount']].plot()
questions.sort_values(by='FavCount').groupby('FavCount').median()[['ViewCount']].plot()
questions.sort_values(by='CommentCount').groupby('CommentCount').median()[['ViewCount']].plot()
```

## Correlation between covariates.

```python
questions.sort_values(by='AnsCount').groupby('AnsCount').median()[['CommentCount','FavCount','ViewAge']].plot()
questions.sort_values(by='FavCount').groupby('FavCount').median().plot(y='CommentCount')
#questions.sort_values(by='age').groupby('age').median().plot(y='ViewCount')
```

NB: Is there a correlation between favCount and CommentCount?

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
mod_ols = smf.ols(formula='np.log(ViewAge + 1) ~ Score + CommentCount + AnsCount + FavCount - 1', data=questions)
res = mod_ols.fit()
print res.summary()
```

```python
mod_poi = smf.glm('ViewCount ~ Score + AnsCount + FavCount + CommentCount', data=questions, offset=np.log(questions['age']),
                  family=sm.families.Poisson(link=sm.families.links.log)).fit(method='bfgs')
#print mod_poi.mle_retvals['converged']
print mod_poi.summary()

```

```python
huber_t = sm.RLM(np.log(questions.ViewCount + 1), questions[['Score','FavCount']], M=sm.robust.norms.HuberT())
hub_results = huber_t.fit()

print hub_results.params
print hub_results.bse
#varnames = ['var_%d' % i for i in range(len(hub_results.params))]
print hub_results.summary()
```
