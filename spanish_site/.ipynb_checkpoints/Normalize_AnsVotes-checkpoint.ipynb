```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
votes = pd.DataFrame.from_csv('AnsVotes_TSeries.csv', index_col=None)
votes.head(10)
```

```python
votes[['Age_vot','AnsWordCount','Comments','Ans_count']].describe()
```

```python
skewed_vars_pos = ['Comments','AnsWordCount','QVotes','Votes','CumVotes']
skewed_vars_neg = ['DRank','QScore','Score','CumScore']

for vname in skewed_vars_pos:
    votes['log_' + vname] = np.log(votes[vname] + 1.0)
for vname in skewed_vars_neg:
    votes['log_' + vname] = np.log(votes[vname] - votes[vname].min() + 1.0)
```

```python
rvars = ['Pbias','EPbias','ReScore','QReScore']   
       
for vname in rvars:
    votes[vname]=(votes[vname] - votes[vname].mean())/votes[vname].std()
```

```python
#Normalize interactions
to_interact = ['log_Comments','log_AnsWordCount','log_Score']

votes['Social_signal'] = votes['log_Comments'] + votes['log_AnsWordCount'] + votes['log_Score']
inter = votes['ReScore'] * votes['EPbias'] * votes['Social_signal']
inter = (inter - inter.mean()) / inter.std()
votes["ReScore" + "_x_" + "EPbias"+ "_x_" + 'Social_signal'] = inter

for vname in to_interact:
    x = votes['ReScore'] * votes['EPbias'] * votes[vname]
    x = (x - x.mean()) / x.std()
    votes["ReScore" + "_x_" + "EPbias" +  "_x_" + vname] = x
```

```python
votes.to_csv("AnsVotes_TSeries_normalized.csv", float_format="%.4f")
```
